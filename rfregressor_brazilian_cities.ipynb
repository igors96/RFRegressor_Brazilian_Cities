```python
import pandas as pd
pd.set_option('display.max_columns', 81)
cidades = pd.read_csv('/content/BRAZIL_CITIES.csv', sep = ';', decimal = ',')
```

```python
cidades.head()
```

```python
cidades.drop(['CITY', 'IDHM Ranking 2010', 'IDHM_Renda', 'IDHM_Longevidade', 'IDHM_Educacao', 'LONG', 'LAT', 'GVA_MAIN', 'REGIAO_TUR', 'MUN_EXPENDIT', 'HOTELS', 'BEDS', 'Pr_Agencies', 'Pu_Agencies', 'Pr_Bank', 'Pu_Bank', 'Pr_Assets', 'Pu_Assets','UBER','MAC','WAL-MART'], axis=1, inplace=True)
```

```python
cidades.head()
```

```python
state_encode = pd.get_dummies(cidades['STATE'])
rural_encode = pd.get_dummies(cidades['RURAL_URBAN'])
categoria_encode = pd.get_dummies(cidades['CATEGORIA_TUR'])
cidades.drop(['STATE', 'RURAL_URBAN', 'CATEGORIA_TUR'], axis = 1, inplace = True)

concatenado = pd.concat([cidades, state_encode, rural_encode, categoria_encode], axis = 1)
```

```python
concatenado.head()
```

```python
novo = concatenado.dropna()
```

```python
novo['AREA']= novo['AREA'].str.replace(',', '')
```

```python
x = novo.drop('IDHM', axis = 1)
y = novo['IDHM']
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

normalizador = MinMaxScaler(feature_range = (0,1))
x_norm = normalizador.fit_transform(x)

pca = PCA(n_components = 15)
x_pca = pca.fit_transform(x_norm)
```

```python
print('Variância explicada dos componentes:', pca.explained_variance_ratio_)
```

```python
sum(pca.explained_variance_ratio_)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

modelo = RandomForestRegressor(n_estimators = 100, n_jobs = -1)
kfold = KFold(n_splits = 3)
resultado = cross_val_score(modelo, x_pca, y, cv = kfold, n_jobs = -1)
print(resultado.mean())
```

```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV

minimos_split = np.array([2,3,4])
maximo_nivel = np.array([3,5,7,9,11,14])
minimo_leaf = np.array([3,4,5,6])
valores_grid = {'min_samples_split': minimos_split, 'max_depth': maximo_nivel, 'min_samples_leaf': minimo_leaf}

modelo1 = RandomForestRegressor(n_estimators = 100, n_jobs = -1)

gridRandomForest = RandomizedSearchCV(estimator = modelo1, param_distributions = valores_grid, cv = 3, n_iter = 50, n_jobs = -1)
gridRandomForest.fit(x_pca,y)

print('Mínimo split: ', gridRandomForest.best_estimator_.min_samples_split)
print('Máxima profundidade: ', gridRandomForest.best_estimator_.max_depth)
print('Mínimo leaf: ', gridRandomForest.best_estimator_.min_samples_leaf)
print('R2: ', gridRandomForest.best_score_)
```
